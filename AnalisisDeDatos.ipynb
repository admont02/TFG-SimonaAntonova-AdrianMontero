```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
#Cargar el archivo JSON
with open('AnalisisPruebas/cuestionario1.json', 'r') as f:
    raw_data = json.load(f)

#Extraer cada usuario como una fila en el DataFrame
data_list = []
for user_token, user_data in raw_data.items():
    if user_data is not None:  # Validar que los datos del usuario no sean None
        #Agregar el token como una columna separada
        user_data["token"] = user_token.split('_')[-1][:4]
        #Añadir al listado de datos
        data_list.append(user_data)

#Crear el DataFrame a partir de la lista de usuarios
df = pd.DataFrame(data_list)

#Mostrar todas las columnas disponibles (sin recortes visuales en Jupyter)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#Mostrar todo el DataFrame (cada fila corresponde a un usuario)
#display(df)

#Filtrar únicamente las columnas relevantes
columnas_interes = [
    'token', 'G01Q04',
    'G01Q05[SQ01]', 'G01Q05[SQ02]', 'G01Q05[SQ03]', 'G01Q05[SQ04]'
    ,'G02Q01','G02Q02','G02Q03','G02Q04','G02Q05','G02Q06', 
    'G02Q07','G02Q08','G02Q09','G02Q10','G02Q11','G02Q12',
    'G02Q13','G02Q14','G02Q15','G02Q16','G02Q17',
    'G02Q18','G02Q19','G02Q20','id'
]
df_filtrado = df[columnas_interes]

#Mostrar el DataFrame filtrado
display(df_filtrado)


```

```python
#Calcular la edad media
edad_media = df['G01Q01'].astype(float).mean()

#Gráfico de la edad media
plt.figure(figsize=(4, 3))
sns.barplot(x=["Edad Media"], y=[edad_media], palette="coolwarm")
plt.title("Edad Media de los Usuarios", fontsize=14)
plt.ylabel("Edad", fontsize=12)
plt.xlabel("")
plt.show()
print(f"Edad media de los usuarios: {edad_media:.2f}")
```

```python
#Contar respuestas de género (G01Q02)
genero = df['G01Q02'].value_counts()
mujeres = genero.get('AO01', 0)  #Mujeres (AO01)
hombres = genero.get('AO02', 0)  #Hombres (AO02)
#Gráfico de barras para el género
plt.figure(figsize=(6, 4))
sns.barplot(x=genero.index, y=genero.values, palette="viridis")
plt.title("Distribución de Género", fontsize=14)
plt.xlabel("Género", fontsize=12)
plt.ylabel("Frecuencia", fontsize=12)
plt.xticks(ticks=[0, 1], labels=["Mujer (AO01)", "Hombre (AO02)"])
plt.show()
print(f"{mujeres} mujeres y {hombres} hombres")
```
