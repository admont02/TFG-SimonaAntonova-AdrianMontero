```python
# Diccionario con respuestas correctas
respuestas_correctas = {
    "G02Q01": "AO01", "G02Q02": "AO01", "G02Q03": "AO03", "G02Q04": "AO02", 
    "G02Q05": "AO01", "G02Q06": "AO03", "G02Q07": "AO03", "G02Q08": "AO01",
    "G02Q09": "AO01", "G02Q10": "AO03", "G02Q11": "AO03", "G02Q12": "AO01",
    "G02Q13": "AO03", "G02Q14": "AO03", "G02Q15": "AO03", "G02Q16": "AO02",
    "G02Q17": "AO01", "G02Q18": "AO03", "G02Q19": "AO01", "G02Q20": "AO01"
}

# Calcular aciertos y fallos para el pre-test
df_pre["aciertos_pre"] = df_pre[preguntas_G02].apply(
    lambda row: sum(row[q] == respuestas_correctas[q] for q in preguntas_G02), axis=1
)
df_pre["fallos_pre"] = len(preguntas_G02) - df_pre["aciertos_pre"]

# Identificar preguntas falladas en el pre-test
df_pre["preguntas_falladas_pre"] = df_pre[preguntas_G02].apply(
    lambda row: [q for q in preguntas_G02 if row[q] != respuestas_correctas[q]], axis=1
)

# Calcular aciertos y fallos para el post-test
df_post["aciertos_post"] = df_post[preguntas_G02].apply(
    lambda row: sum(row[q] == respuestas_correctas[q] for q in preguntas_G02), axis=1
)
df_post["fallos_post"] = len(preguntas_G02) - df_post["aciertos_post"]

# Identificar preguntas falladas en el post-test
df_post["preguntas_falladas_post"] = df_post[preguntas_G02].apply(
    lambda row: [q for q in preguntas_G02 if row[q] != respuestas_correctas[q]], axis=1
)

# Agregar la edad al DataFrame de resultados desde el pre-test
df_resultados = pd.merge(
    df_pre[["token", "G01Q01", "aciertos_pre", "fallos_pre", "preguntas_falladas_pre"]],
    df_post[["token", "aciertos_post", "fallos_post", "preguntas_falladas_post"]],
    on="token"
)

# Renombrar la columna G01Q01 a "edad" para mayor claridad
df_resultados = df_resultados.rename(columns={"G01Q01": "edad"})
pd.set_option('display.max_colwidth', None)
# Mostrar el DataFrame actualizado con la edad y las preguntas falladas
display(df_resultados)
```

# Análisis de Trazas del Juego

```python
from collections import defaultdict
from datetime import datetime
import json

# Cargar el archivo de trazas
with open("AnalisisPruebas/traces.json", "r", encoding="utf-8") as f:
    trazas = json.load(f)

# Usuarios de prueba que queremos eliminar
usuarios_a_eliminar = ["67f52f0490d7c302b9aebce2_jtor", "67f52f0490d7c302b9aebce2_maei", 
                       "67f52f0490d7c302b9aebce2_iwhu", "67f52f0490d7c302b9aebce2_uyit", 
                       "67f52f0490d7c302b9aebce2_avwl", "67f52f0490d7c302b9aebce2_vrtd", 
                       "67f52f0490d7c302b9aebce2_wnnk","67f52f0490d7c302b9aebce2_zbub"]

# Filtrar trazas eliminando usuarios de prueba
trazas = [traza for traza in trazas if traza["actor"]["name"] not in usuarios_a_eliminar]

# Diccionario para eventos separados por usuario
eventos_por_usuario = defaultdict(list)

# Diccionario para almacenar múltiples intentos por nivel y usuario
intentos_por_nivel = defaultdict(lambda: defaultdict(list))

# Diccionario para contar movimientos por nivel y usuario
movimientos_por_nivel = defaultdict(lambda: defaultdict(lambda: defaultdict(int)))

# Procesar las trazas
for traza in trazas:
    usuario = traza["actor"]["name"] if "actor" in traza and "name" in traza["actor"] else "Desconocido"
    nivel = traza["object"]["id"] if "object" in traza and "id" in traza["object"] else "Sin Nivel"
    accion = traza["verb"]["display"]["en"] if "verb" in traza and "display" in traza["verb"] else "Acción Desconocida"
    timestamp = traza["timestamp"] if "timestamp" in traza else None
    success = traza["result"]["success"] if accion == "completed" and "result" in traza and "success" in traza["result"] else None

    # Convertir el timestamp a objeto datetime si existe
    timestamp_dt = datetime.fromisoformat(timestamp.replace("Z", "")) if timestamp else None

    # Crear un evento único
    evento = {
        "Nivel": nivel,
        "Acción": accion,
        "Timestamp": timestamp_dt,
        "Success": success
    }
    
    # Añadir el evento a la lista del usuario
    eventos_por_usuario[usuario].append(evento)

    # Calcular intentos para niveles principales
    if accion == "initialized":
        # Añadir un nuevo intento con su marca de inicio
        intentos_por_nivel[usuario][nivel].append({"initialized": timestamp_dt, "completed": None, "duracion": None,
            "success": None })
    elif accion == "completed" and nivel in intentos_por_nivel[usuario]:
        # Buscar el intento más reciente sin completar y actualizarlo
        for intento in intentos_por_nivel[usuario][nivel]:
            if intento["completed"] is None:
                intento["completed"] = timestamp_dt
                intento["success"] = success
                # Calcular duración si ambas marcas están presentes
                if intento["initialized"] and intento["completed"]:
                    intento["duracion"] = (intento["completed"] - intento["initialized"]).total_seconds()
                break  # Salimos porque solo actualizamos el intento más reciente

    # Contar movimientos específicos asociados al nivel
    if "player-moved-" in nivel:
        tipo_movimiento = nivel.split("/")[-1].replace("player-moved-", "")
        movimientos_por_nivel[usuario][nivel][tipo_movimiento] += 1


```

### A continuación, aparecen tres tablas por cada uno de los usuarios, separadas por los tipos de nivel: Prioridad,Luces y Manejo

```python
# Agrupar niveles en distintas categorías
grupo1 = ["nivel3", "nivel12", "nivel13", "nivel18"]
grupo2 = ["nivel8", "nivel9", "nivel10", "nivel11"]

# Columnas específicas para cada grupo
columns_grupo1 = ["Intento", "Nivel", "Resultado", "Tiempo (s)", "Antinieblas Traseras ON", "Cortas ON", 
                  "Antinieblas Delanteras ON", "Largas ON", "Forward", "Right", "Left", "Backward"]
columns_grupo2 = ["Intento", "Nivel", "Resultado", "Tiempo (s)"]
columns_otros = ["Intento", "Nivel", "Resultado", "Tiempo (s)", "Forward", "Right", "Left", "Backward"]

# Generar tablas por usuario
print("=== Tablas por Usuario ===")
for usuario, niveles in intentos_por_nivel.items():
    print(f"Usuario: {usuario}")
    data = []
    for nivel, intentos in niveles.items():
        for intento_num, intento in enumerate(intentos, start=1):
            # Inicializar contadores para movimientos y luces
            forward_count, right_count, left_count, backward_count = 0, 0, 0, 0
            antinieblas_traseras_on, antinieblas_traseras_off = 0, 0
            antinieblas_delanteras_on, antinieblas_delanteras_off = 0, 0
            cortas_on, largas_on, cortas_off, largas_off = 0, 0, 0, 0

            # Procesar trazas para acumular movimientos y luces dentro del rango de tiempo del intento
            for traza in trazas:
                traza_usuario = traza["actor"]["name"]
                traza_nivel = traza["object"]["id"]
                traza_timestamp = datetime.fromisoformat(traza["timestamp"].replace("Z", ""))
                
                # Verificar si la traza pertenece al usuario y está dentro del rango de tiempo del intento actual
                if traza_usuario == usuario and intento["initialized"] and intento["completed"]:
                    if intento["initialized"] <= traza_timestamp <= intento["completed"]:
                        # Contar movimientos específicos (player-moved-forward, etc.)
                        if "player-moved-forward" in traza_nivel:
                            forward_count += 1
                        elif "player-moved-right" in traza_nivel:
                            right_count += 1
                        elif "player-moved-left" in traza_nivel:
                            left_count += 1
                        elif "player-moved-backward" in traza_nivel:
                            backward_count += 1
                        # Contar acciones de luces
                        elif "antinieblasTraseras-on" in traza_nivel:
                            antinieblas_traseras_on += 1
                        elif "antinieblasTraseras-off" in traza_nivel:
                            antinieblas_traseras_off += 1
                        elif "antinieblasDelanteras-on" in traza_nivel:
                            antinieblas_delanteras_on += 1
                        elif "antinieblasDelanteras-off" in traza_nivel:
                            antinieblas_delanteras_off += 1
                        elif "cortas-on" in traza_nivel:
                            cortas_on += 1
                        elif "largas-on" in traza_nivel:
                            largas_on += 1
                        elif "cortas-off" in traza_nivel:
                            cortas_off += 1
                        elif "largas-off" in traza_nivel:
                            largas_off += 1

            # Crear fila para el DataFrame
            data.append({
                "Intento": intento_num,
                "Nivel": nivel.split("/")[-1],  # Simplificar el nombre del nivel
                "Resultado": "Éxito" if intento["success"] else "Fallo",
                "Tiempo (s)": intento["duracion"] if intento["duracion"] else "Sin datos",
                "Forward": forward_count,
                "Right": right_count,
                "Left": left_count,
                "Backward": backward_count,
                "Antinieblas Traseras ON": antinieblas_traseras_on,
                "Antinieblas Traseras OFF": antinieblas_traseras_off,
                "Antinieblas Delanteras ON": antinieblas_delanteras_on,
                "Antinieblas Delanteras OFF": antinieblas_delanteras_off,
                "Cortas ON": cortas_on,
                "Largas ON": largas_on,
                "Cortas OFF": cortas_off,
                "Largas OFF": largas_off
            })

    # Crear DataFrame general
    df = pd.DataFrame(data)

    # Eliminar filas donde "Tiempo (s)" sea "Sin datos"
    df = df[df["Tiempo (s)"] != "Sin datos"]

    # Crear tabla para cada grupo de niveles
    grupo1_df = df[df["Nivel"].isin(grupo1)][columns_grupo1]  # Filtrar columnas específicas para Grupo 1
    grupo2_df = df[df["Nivel"].isin(grupo2)][columns_grupo2]  # Filtrar columnas específicas para Grupo 2
    otros_df = df[~df["Nivel"].isin(grupo1 + grupo2)][columns_otros]  # Filtrar columnas específicas para Otros

    # Mostrar las tablas
    print("=== Niveles Luces ===")
    display(grupo1_df)

    print("=== Niveles Prioridad ===")
    display(grupo2_df)

    print("=== Niveles Manejo ===")
    display(otros_df)

    print("\n")
```
