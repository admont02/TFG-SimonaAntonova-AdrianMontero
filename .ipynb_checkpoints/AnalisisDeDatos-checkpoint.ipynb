```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import seaborn as sns
#Cargar el archivo JSON
with open('AnalisisPruebas/cuestionario1.json', 'r') as f:
    raw_data = json.load(f)

#Extraer cada usuario como una fila en el DataFrame
data_list = []
for user_token, user_data in raw_data.items():
    if user_data is not None:  # Validar que los datos del usuario no sean None
        #Agregar el token como una columna separada
        user_data["token"] = user_token.split('_')[-1][:4]
        #Añadir al listado de datos
        data_list.append(user_data)

#Crear el DataFrame a partir de la lista de usuarios
df = pd.DataFrame(data_list)

#Mostrar todas las columnas disponibles (sin recortes visuales en Jupyter)
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

#Mostrar todo el DataFrame (cada fila corresponde a un usuario)
#display(df)

#Filtrar únicamente las columnas relevantes
columnas_interes = [
    'token', 'G01Q04',
    'G01Q05[SQ01]', 'G01Q05[SQ02]', 'G01Q05[SQ03]', 'G01Q05[SQ04]'
    ,'G02Q01','G02Q02','G02Q03','G02Q04','G02Q05','G02Q06', 
    'G02Q07','G02Q08','G02Q09','G02Q10','G02Q11','G02Q12',
    'G02Q13','G02Q14','G02Q15','G02Q16','G02Q17',
    'G02Q18','G02Q19','G02Q20','id'
]
df_filtrado = df[columnas_interes]

#Mostrar el DataFrame filtrado
display(df_filtrado)


```

```python
import pandas as pd
import json

# Cargar los archivos JSON
with open('AnalisisPruebas/cuestionario1.json', 'r') as f_pre:
    raw_data_pre = json.load(f_pre)

with open('AnalisisPruebas/cuestionario2.json', 'r') as f_post:
    raw_data_post = json.load(f_post)

# Convertir datos en DataFrames para pre y post
data_list_pre = []
data_list_post = []

for user_token, user_data in raw_data_pre.items():
    if user_data is not None:
        user_data["token"] = user_token[-4:]  # Usar últimos 4 caracteres del token
        data_list_pre.append(user_data)

for user_token, user_data in raw_data_post.items():
    if user_data is not None:
        user_data["token"] = user_token[-4:]  # Usar últimos 4 caracteres del token
        data_list_post.append(user_data)

# Crear DataFrames
df_pre = pd.DataFrame(data_list_pre)
df_post = pd.DataFrame(data_list_post)

# Filtrar las columnas que comienzan con G02
preguntas_G02 = [col for col in df_pre.columns if col.startswith("G02")]

# Filtrar las preguntas relevantes en ambos DataFrames
df_pre = df_pre[["token","G01Q01"] + preguntas_G02]
df_post = df_post[["token"] + preguntas_G02]
if "G02Q16" in df_pre.columns:
    df_pre = df_pre.drop(columns=["G02Q16"])

if "G02Q16" in df_post.columns:
    df_post = df_post.drop(columns=["G02Q16"])

# Eliminar también de la lista de preguntas G02
preguntas_G02 = [col for col in preguntas_G02 if col != "G02Q16"]
# Mostrar los DataFrames filtrados
display(df_pre)
display(df_post)
```

```python
# Diccionario con respuestas correctas
respuestas_correctas = {
    "G02Q01": "AO01", "G02Q02": "AO01", "G02Q03": "AO03", "G02Q04": "AO02", 
    "G02Q05": "AO01", "G02Q06": "AO03", "G02Q07": "AO03", "G02Q08": "AO01",
    "G02Q09": "AO01", "G02Q10": "AO03", "G02Q11": "AO03", "G02Q12": "AO01",
    "G02Q13": "AO03", "G02Q14": "AO03", "G02Q15": "AO03", "G02Q16": "AO02",
    "G02Q17": "AO01", "G02Q18": "AO03", "G02Q19": "AO01", "G02Q20": "AO01"
}

# Calcular aciertos y fallos para el pre-test
df_pre["aciertos_pre"] = df_pre[preguntas_G02].apply(
    lambda row: sum(row[q] == respuestas_correctas[q] for q in preguntas_G02), axis=1
)
df_pre["fallos_pre"] = len(preguntas_G02) - df_pre["aciertos_pre"]

# Calcular aciertos y fallos para el post-test
df_post["aciertos_post"] = df_post[preguntas_G02].apply(
    lambda row: sum(row[q] == respuestas_correctas[q] for q in preguntas_G02), axis=1
)
df_post["fallos_post"] = len(preguntas_G02) - df_post["aciertos_post"]

# Mostrar resultados con nuevas columnas
display(df_pre)
display(df_post)
# Agregar la edad al DataFrame de resultados desde el pre-test
df_resultados = pd.merge(
    df_pre[["token","G01Q01", "aciertos_pre"]],  # Incluimos la columna G01Q01 (edad)
    df_post[["token", "aciertos_post"]],
    on="token"
)

# Renombrar la columna G01Q01 a "edad" para mayor claridad
df_resultados = df_resultados.rename(columns={"G01Q01": "edad"})

# Mostrar el DataFrame actualizado con la edad
display(df_resultados)
```

```python
# Estadísticas globales
estadisticas_globales = {
    "Pre-Test (Aciertos)": {
        "Media": df_pre["aciertos_pre"].mean(),
        "Desviación Estándar": df_pre["aciertos_pre"].std(),
        "Distribución": df_pre["aciertos_pre"].value_counts()
    },
    "Post-Test (Aciertos)": {
        "Media": df_post["aciertos_post"].mean(),
        "Desviación Estándar": df_post["aciertos_post"].std(),
        "Distribución": df_post["aciertos_post"].value_counts()
    }
}

# Mostrar estadísticas globales
for test, stats in estadisticas_globales.items():
    print(f"\n{test}")
    for key, value in stats.items():
        print(f"{key}: {value}")
```

```python
# Comparar aciertos por pregunta (media por columna)
mejoras_por_pregunta = pd.DataFrame({
    "Media Pre-Test": [df_pre[q].apply(lambda x: x == respuestas_correctas[q]).mean() for q in preguntas_G02],
    "Media Post-Test": [df_post[q].apply(lambda x: x == respuestas_correctas[q]).mean() for q in preguntas_G02]
}, index=preguntas_G02)

mejoras_por_pregunta["Diferencia"] = mejoras_por_pregunta["Media Post-Test"] - mejoras_por_pregunta["Media Pre-Test"]

# Mostrar mejoras por pregunta
display(mejoras_por_pregunta)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Visualizar mejoras globales (aciertos pre y post)
plt.figure(figsize=(8, 5))
sns.barplot(x=["Pre-Test", "Post-Test"], y=[
    estadisticas_globales["Pre-Test (Aciertos)"]["Media"],
    estadisticas_globales["Post-Test (Aciertos)"]["Media"]
], palette="coolwarm")
plt.title("Media de Aciertos: Pre vs Post", fontsize=14)
plt.ylabel("Media de Aciertos", fontsize=12)
plt.xlabel("")
plt.show()

# Visualizar mejoras por pregunta
plt.figure(figsize=(12, 6))
sns.barplot(x=mejoras_por_pregunta.index, y=mejoras_por_pregunta["Diferencia"], palette="viridis")
plt.xticks(rotation=45)
plt.title("Diferencia de Media de Aciertos por Pregunta", fontsize=14)
plt.ylabel("Diferencia en Media", fontsize=12)
plt.xlabel("Preguntas", fontsize=12)
plt.show()
```

```python
import json
from collections import Counter
import pandas as pd
with open("AnalisisPruebas/traces.json", "r", encoding="utf-8") as f:
    trazas = json.load(f)

    
from collections import defaultdict

# Crear un diccionario para agrupar trazas por usuario
trazas_por_usuario = defaultdict(list)

# Agrupar las trazas
for traza in trazas:
    usuario = traza["actor"]["name"] if "actor" in traza and "name" in traza["actor"] else "Desconocido"
    trazas_por_usuario[usuario].append(traza)

# Mostrar los usuarios y el número de trazas por cada uno
for usuario, trazas_usuario in trazas_por_usuario.items():
    print(f"Usuario: {usuario}, Número de trazas: {len(trazas_usuario)}")
    
    
    
# Análisis de acciones (verbs)
verbs = [traza["verb"]["display"]["en"] for traza in trazas if "verb" in traza]
conteo_verbs = Counter(verbs)
display("Acciones más comunes:", conteo_verbs)



# Análisis de objetos (activity objects)
#objetos = [traza["object"]["id"] for traza in trazas if "object" in traza]
#conteo_objetos = Counter(objetos)
#display("Objetos más utilizados:", conteo_objetos)
```

```python
from collections import defaultdict
from datetime import datetime
with open("AnalisisPruebas/traces.json", "r", encoding="utf-8") as f:
    trazas = json.load(f)
usuarios_a_eliminar = ["67f52f0490d7c302b9aebce2_jtor", "67f52f0490d7c302b9aebce2_maei","67f52f0490d7c302b9aebce2_iwhu"
                      ,"67f52f0490d7c302b9aebce2_uyit","67f52f0490d7c302b9aebce2_avwl","67f52f0490d7c302b9aebce2_vrtd","67f52f0490d7c302b9aebce2_wnnk"]
# Diccionario para eventos separados por usuario
eventos_por_usuario = defaultdict(list)

# Diccionario para almacenar múltiples intentos por nivel y usuario
intentos_por_nivel = defaultdict(lambda: defaultdict(list))
trazas = [traza for traza in trazas if traza["actor"]["name"] not in usuarios_a_eliminar]
# Crear eventos separados por nivel y verbo
for traza in trazas:
    usuario = traza["actor"]["name"] if "actor" in traza and "name" in traza["actor"] else "Desconocido"
    nivel = traza["object"]["id"] if "object" in traza and "id" in traza["object"] else "Sin Nivel"
    accion = traza["verb"]["display"]["en"] if "verb" in traza and "display" in traza["verb"] else "Acción Desconocida"
    timestamp = traza["timestamp"] if "timestamp" in traza else None
    success = traza["result"]["success"] if accion == "completed" and "result" in traza and "success" in traza["result"] else None

    # Convertir el timestamp a objeto datetime si existe
    timestamp_dt = datetime.fromisoformat(timestamp.replace("Z", "")) if timestamp else None

    # Crear un evento único
    evento = {
        "Nivel": nivel,
        "Acción": accion,
        "Timestamp": timestamp_dt,
        "Success": success
    }
    
    # Añadir el evento a la lista del usuario
    eventos_por_usuario[usuario].append(evento)

    # Calcular intentos para niveles
    if accion == "initialized":
        # Añadir un nuevo intento con su marca de inicio
        intentos_por_nivel[usuario][nivel].append({"initialized": timestamp_dt, "completed": None, "duracion": None})
    elif accion == "completed" and nivel in intentos_por_nivel[usuario]:
        # Buscar el intento más reciente sin completar y actualizarlo
        for intento in intentos_por_nivel[usuario][nivel]:
            if intento["completed"] is None:
                intento["completed"] = timestamp_dt
                # Calcular duración si ambas marcas están presentes
                if intento["initialized"] and intento["completed"]:
                    intento["duracion"] = (intento["completed"] - intento["initialized"]).total_seconds()
                break  # Salimos porque solo actualizamos el intento más reciente

# Mostrar resultados organizados
print("=== Eventos Separados por Usuario ===")
for usuario, eventos in eventos_por_usuario.items():
    print(f"Usuario: {usuario}")
    for evento in eventos:
        if evento["Acción"] == "completed" and evento["Nivel"] in intentos_por_nivel[usuario]:
            # Obtener la duración del intento más reciente completado
            intentos = intentos_por_nivel[usuario][evento["Nivel"]]
            ultimo_intento = next((i for i in intentos if i["completed"] == evento["Timestamp"]), None)
            duracion = ultimo_intento["duracion"] if ultimo_intento else "Sin datos"
            success_info = f", Success - {evento['Success']}, Tiempo - {duracion}s"
            print(f"  Evento: Nivel - {evento['Nivel']}, Acción - {evento['Acción']}{success_info}")
        else:
            print(f"  Evento: Nivel - {evento['Nivel']}, Acción - {evento['Acción']}")
```
